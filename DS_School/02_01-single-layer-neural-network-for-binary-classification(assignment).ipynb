```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### Generate Dataset

```python
x1 = np.random.uniform(low=-1.0, high=1.0, size=100)

print(x1.shape)
x1[0:10]
```

```python
x2 = np.random.uniform(low=-1.0, high=1.0, size=100)

print(x2.shape)
x2[0:10]
```

```python
# y =(y>0) 을통하여 0이냐 1이냐를 알아봄. int로 안바꾸면 true and false 형식으로 나옴

y = 0.3 * x1 + 0.5 * x2 + 0.1

y = (y > 0).astype(int)


print(y.shape)
y[0:10]
```

```python
plt.scatter(x1, x2, c=y)
```

## Define Sigmoid

```python
def sigmoid(n):
    return 1 / (1 + np.exp(-n))

xx = np.linspace(start=-10, stop=+10)
yy = sigmoid
```

### Gradient Descent

```python
num_epoch = 100
learning_rate = 1.0

w1 = np.random.uniform(low=-1.0, high=1.0)
w2 = np.random.uniform(low=-1.0, high=1.0)
b = np.random.uniform(low=-1.0, high=1.0)

for epoch in range(num_epoch):
    y_predict = w1 * x1 + w2 * x2 + b
    y_predict = sigmoid(y_predict)
    
    predict = (y_predict > 0.5).astype('int')
    accuracy = (predict == y).mean()
    
    print("{0:2} accuracy = {1:.6f}".format(epoch, accuracy))
    
   
    if accuracy >= 0.95:
        break

    w1 = w1 - learning_rate * ((y_predict - y) * x1).mean()
    w2 = w2 - learning_rate * ((y_predict - y) * x2).mean()
    b = b - learning_rate * (y_predict - y).mean()

print("----" * 10)
print("{0:2} accuracy = {1:.6f}".format(epoch, accuracy))
```

```python
import pandas as pd

y_predict = w1 * x1 + w2 * x2 + b
y_predict = sigmoid(y_predict)
predict = (y_predict > 0.5).astype('int')

result = pd.DataFrame({'x1': x1, 'x2': x2, 'y(actual)': y, 'y(predict)': predict})

result
```
